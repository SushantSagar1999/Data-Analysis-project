# Gender Distribution

```python
 plt.figure(figsize = (5, 5))
 sns.countplot(data = df, x = "Gender")
 plt.title("Gender Distribution")
 plt.show()
```

from the above chart we have analysed that the number of female in data is 
more than number than male.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',
                                   "ReadingScore":'mean',
                                  "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between ParentEduc and Student Score ")
plt.show()
```

from the above chat we have the education of the parents have a good impact on their students marks.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',
                                   "ReadingScore":'mean',
                                  "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between ParentMaritalStatus and Student Score ")
plt.show()
```

from the upper chart we have seen the ParentMaritalStatus does not effect the student result weather they are live single, widowed or they Divorce.

# Distribution of Ethnic Group:

```python
groupA = df.loc[(df['EthnicGroup']== "group A")].count()
groupB = df.loc[(df['EthnicGroup']== "group B")].count()
groupC = df.loc[(df['EthnicGroup']== "group C")].count()
groupD = df.loc[(df['EthnicGroup']== "group D")].count()
groupE = df.loc[(df['EthnicGroup']== "group E")].count()

mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"],
        groupD["EthnicGroup"], groupE["EthnicGroup"]]
Labels = ["groupA", "groupB", "groupC", "groupD", "groupE" ]

plt.pie(mlist, labels = Labels, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.legend(Labels, title="Ethnic Groups", loc = "upper right")
plt.show();
```
